# Customer Intelligence Platform
**Dataset:** UCI Online Retail (541,909 real transactions, UK e-commerce, 2010–2011)  
**Goal:** Segment customers by value, find product bundles, predict churn, deploy dashboard

---

## Step 1 — Load the Dataset

```python
import sys
sys.path.insert(0, 'src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from data_loader import load_online_retail

raw = load_online_retail()
print(f'Shape: {raw.shape}')
raw.head()
```

## Step 2 — Clean the Data

```python
from cleaning import clean_online_retail

df = clean_online_retail(raw)
print(f'Clean shape: {df.shape}')
df.head()
```

## Step 3 — Exploratory Data Analysis (EDA)

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 4))

# Revenue by country (top 10)
top_countries = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)
top_countries.plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Revenue by Country (Top 10)')
axes[0].set_ylabel('Revenue (£)')
axes[0].tick_params(axis='x', rotation=45)

# Monthly revenue
df['Month'] = df['InvoiceDate'].dt.to_period('M')
monthly = df.groupby('Month')['TotalPrice'].sum()
monthly.plot(ax=axes[1], color='green', marker='o')
axes[1].set_title('Monthly Revenue')
axes[1].set_ylabel('Revenue (£)')
axes[1].tick_params(axis='x', rotation=45)

# Top 10 products by revenue
top_products = df.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='barh', ax=axes[2], color='coral')
axes[2].set_title('Top 10 Products by Revenue')
axes[2].set_xlabel('Revenue (£)')

plt.tight_layout()
plt.show()

print(f'Total Revenue:     £{df["TotalPrice"].sum():,.0f}')
print(f'Unique Customers:  {df["CustomerID"].nunique():,}')
print(f'Unique Orders:     {df["InvoiceNo"].nunique():,}')
print(f'Date Range:        {df["InvoiceDate"].min().date()} → {df["InvoiceDate"].max().date()}')
```

## Step 4 — RFM Segmentation + KMeans Clustering

```python
from rfm_segmentation import compute_rfm, add_rfm_scores, cluster_customers

rfm = compute_rfm(df)
rfm = add_rfm_scores(rfm)
rfm = cluster_customers(rfm, n_clusters=4)

print('Segment sizes:')
print(rfm['Segment'].value_counts())
print()
print('Mean R/F/M per segment:')
rfm.groupby('Segment')[['Recency','Frequency','Monetary']].mean().round(1)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Scatter: Recency vs Monetary
colors = {'Champions':'gold','Loyal Customers':'steelblue',
          'Potential Loyalists':'green','At Risk':'red','Hibernating':'grey'}
for seg, group in rfm.groupby('Segment'):
    axes[0].scatter(group['Recency'], group['Monetary'],
                    label=seg, alpha=0.6, s=group['Frequency']*3,
                    color=colors.get(seg, 'purple'))
axes[0].set_xlabel('Recency (days since last purchase)')
axes[0].set_ylabel('Monetary (£ total spent)')
axes[0].set_title('Customer Segments — Recency vs Monetary')
axes[0].legend()

# Bar: segment sizes
rfm['Segment'].value_counts().plot(kind='bar', ax=axes[1], color='steelblue')
axes[1].set_title('Number of Customers per Segment')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()

rfm.to_csv('data/rfm_segments.csv', index=False)
print('Saved → data/rfm_segments.csv')
```

## Step 5 — Market Basket Analysis (Apriori)

```python
from market_basket import run_market_basket

rules = run_market_basket(df, country='United Kingdom', min_support=0.02, min_lift=1.0)
rules.to_csv('data/market_basket_rules.csv', index=False)
print('Saved → data/market_basket_rules.csv')
rules.head(10)
```

```python
# Visualise top 10 rules by lift
top_rules = rules.head(10).copy()
top_rules['rule'] = top_rules['antecedents'] + '  →  ' + top_rules['consequents']

fig, ax = plt.subplots(figsize=(12, 5))
bars = ax.barh(top_rules['rule'], top_rules['lift'], color='coral')
ax.set_xlabel('Lift')
ax.set_title('Top 10 Product Association Rules by Lift')
ax.axvline(x=1, color='black', linestyle='--', label='Lift = 1 (random)')
ax.legend()
plt.tight_layout()
plt.show()
```

## Step 6 — Churn Prediction

```python
from churn_model import build_churn_dataset, train_churn_models

churn_features = build_churn_dataset(df, observation_days=90)
print(f'Customers: {len(churn_features)}')
print(f'Churn rate: {churn_features["Churned"].mean():.1%}')
churn_features.head()
```

```python
models = train_churn_models(churn_features)

# Feature importance chart
rf = models['random_forest']
feat_cols = models['feature_cols']

imp = pd.DataFrame({'Feature': feat_cols, 'Importance': rf.feature_importances_})\
        .sort_values('Importance', ascending=True)

fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(imp['Feature'], imp['Importance'], color='steelblue')
ax.set_title('Random Forest — Feature Importances for Churn Prediction')
ax.set_xlabel('Importance')
plt.tight_layout()
plt.show()

churn_features['ChurnProbability'] = rf.predict_proba(churn_features[feat_cols])[:, 1]
churn_features.to_csv('data/churn_features.csv', index=False)
print('Saved → data/churn_features.csv')

print('\nTop 10 highest churn-risk customers:')
churn_features.sort_values('ChurnProbability', ascending=False)\
    [['CustomerID','Recency','Frequency','Monetary','ChurnProbability']].head(10)
```

## Step 7 — Business Summary

| Finding | Insight |
|---|---|
| Segmentation | Champions (small %) drive majority of revenue — prioritise retention |
| Market Basket | Product pairs with Lift > 1 = bundle/cross-sell opportunities |
| Churn Model | Recency is the strongest churn predictor — customers who haven't bought recently need re-engagement |
